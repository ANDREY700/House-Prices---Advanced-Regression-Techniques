```python
df['BsmtFullBath'].fillna(0, inplace=True) # BsmtFullBath: Basement full bathrooms
df['BsmtHalfBath'].fillna(0, inplace=True) # BsmtHalfBath: Basement half bathrooms
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].median(), inplace=True)
df['MasVnrArea'].fillna(df['MasVnrArea'].median(), inplace=True)
```

```python
df['GarageArea'].fillna(0, inplace=True) # GarageArea: Size of garage in square feet
df['GarageCars'].fillna(0, inplace=True) # GarageCars: Size of garage in car capacity
```

```python
# GarageYrBlt
# df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageYrBlt'] = \
#     df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageYrBlt'].fillna(df['YearBuilt'])

df['GarageYrBlt'].fillna(df['YearBuilt'], inplace=True) # GarageYrBlt: Year garage was built # YearBuilt
df.loc[df['GarageYrBlt'] == 'NAA', 'GarageYrBlt'] = df.loc[df['GarageYrBlt'] == 'NAA', 'YearBuilt'] # YearBuilt
```

```python
df['Alley'].fillna('NA', inplace=True) # Alley: Type of alley access
df['PoolQC'].fillna('NA', inplace=True) # PoolQC: Pool quality
df['Fence'].fillna('NA', inplace=True) # Fence: Fence quality
df['MiscFeature'].fillna('NA', inplace=True) # MiscFeature: Miscellaneous feature not covered in other categories
df['FireplaceQu'].fillna('NA', inplace=True) # FireplaceQu: Fireplace quality
df['SaleType'].fillna('Oth', inplace=True) # SaleType: Type of sale
df['Electrical'].fillna('Mix', inplace=True) # Electrical: Electrical system
df['KitchenQual'].fillna('TA', inplace=True) # KitchenQual: Kitchen quality
```

```python
df.loc[df['Foundation'] == 'Slab', 'BsmtQual'] = 'NA'
df.loc[df['Foundation'] == 'Slab', 'BsmtCond'] = 'NA'
df.loc[df['Foundation'] == 'Slab', 'BsmtExposure'] = 'NA'
df.loc[df['Foundation'] == 'Slab', 'BsmtFinType1'] = 'NA'
df.loc[df['Foundation'] == 'Slab', 'BsmtFinType2'] = 'NA'
```

```python
df.loc[(df['TotalBsmtSF'] == 0.0) & (df['BsmtQual'].isna()), 'BsmtQual'] = 'NA'
df.loc[(df['TotalBsmtSF'] == 0.0) & (df['BsmtCond'].isna()), 'BsmtCond'] = 'NA'
df.loc[(df['TotalBsmtSF'] == 0.0) & (df['BsmtExposure'].isna()), 'BsmtExposure'] = 'NA'
df.loc[(df['TotalBsmtSF'] == 0.0) & (df['BsmtFinType1'].isna()), 'BsmtFinType1'] = 'NA'
df.loc[(df['TotalBsmtSF'] == 0.0) & (df['BsmtFinType2'].isna()), 'BsmtFinType2'] = 'NA'
df.loc[(df['Foundation'] == 'Stone') & (df['BsmtQual'].isna()), 'BsmtQual'] = 'Fa'
df.loc[(df['Foundation'] == 'CBlock') & (df['BsmtCond'].isna()) & (df['BsmtQual'] == 'TA'), 'BsmtCond'] = 'TA' # <------
df.loc[(df['Foundation'] == 'CBlock') & (df['BsmtCond'].isna()) & (df['BsmtQual'] == 'Gd'), 'BsmtCond'] = 'Gd' # <------
df.loc[(df['Foundation'] == 'PConc') & (df['BsmtQual'].isna()) & (df['BsmtCond'] == 'TA'), 'BsmtQual'] = 'TA'  # <------
df.loc[df['BsmtExposure'].isna(),'BsmtExposure'] = 'No' # < ------
df.loc[(df['BsmtQual'] == 'Gd') & (df['BsmtCond'] == 'TA') & (df['BsmtFinType2'].isna()), 'BsmtFinType2'] = 'Unf'
```

```python
df.loc[(df['YearBuilt'] == 1940) & (df['YearRemodAdd'].isin([2005, 2006, 2007, 2008])) & (df['Exterior1st'].isna()), 'Exterior1st'] = 'VinylSd'
df.loc[(df['YearBuilt'] == 1940) & (df['YearRemodAdd'].isin([2005, 2006, 2007, 2008])) & (df['Exterior2nd'].isna()), 'Exterior2nd'] = 'VinylSd'
```

```python
df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageType'] = \
    df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageType'].fillna('NA')

df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageFinish'] = \
    df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageFinish'].fillna('NA')

df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageQual'] = \
    df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageQual'].fillna('NA')

df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageCond'] = \
    df.loc[(df['GarageCars'] == 0.0) & (df['GarageArea'] == 0.0), 'GarageCond'].fillna('NA')
```

```python
df.loc[(df['MasVnrType'].isna()) & (~(df['MasVnrArea'] == 0.0)), 'MasVnrType'] = 'BrkFace'
df.loc[(df['MasVnrType'].isna()) & (df['MasVnrArea'] == 0.0), 'MasVnrType'] = 'No'
```
